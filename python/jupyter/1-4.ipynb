```python
# WordCloud

from wordcloud import WordCloud
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
txt = open('./data/song.txt').read()

mask_img = np.array(Image.open('./data/clover.jpg'))

wc = WordCloud(width=800,
               height=800,
               mask=mask_img,
               background_color='white',
               font_path='./data/H2PORL.ttf').generate(txt)

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
